### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
june_data = df.loc[(df["date"].dt.strftime('%m') == "06"), ['station', 'tobs']] 
print(june_data.head())

dec_data = df.loc[(df["date"].dt.strftime('%m') == "12"), ['station', 'tobs']] 
print(dec_data.head())
```

```python
# Identify the average temperature for June
s_june_avg = june_data.groupby('station').mean()
# s_june_avg
june_avg = june_data['tobs'].mean()
print(f"June Average temp {june_avg}")
```

```python
# Identify the average temperature for December
s_dec_avg = dec_data.groupby('station').mean()
# s_dec_avg
dec_avg = dec_data['tobs'].mean()
dec_avg
print(f"December Average temp {dec_avg}")
```

```python
# Create collections of temperature data
june_temps = june_data['tobs'] 
dec_temps = dec_data['tobs'] 
```

```python
# Run paired t-test
stats.ttest_ind(june_temps,dec_temps)
```
